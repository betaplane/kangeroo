```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
%matplotlib nbagg
```

```python
from IPython.display import set_matplotlib_formats
%matplotlib inline

set_matplotlib_formats('png', 'pdf')
```

```python
x = tf.reshape(tf.linspace(0., 10., 100), (-1, 1))
```

```python
def par(x, a, b, c):
    return tf.concat((tf.zeros_like(x), c * (x - a) * (x - b) / (a - b), c*(x - a), c*(b - x)), 1)
```

```python
tf.InteractiveSession()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
for i, c in enumerate([1., 10.]):
    y = par(x, 3., 6., c)
    axs[i].plot(x.eval(), y.eval()[:, 1:])
    axs[i].plot(x.eval(), tf.nn.softmax(y[:, :2])[:, 1].eval())
    axs[i].grid()
    axs[i].set_ylim(-1, 2)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

for i, b in enumerate([7., 9.]):
    y = par(x, 2., 5., 10.)[:, 1:2]
    z = par(x, 4., b, 10.)[:, 1:2]

    axs[i].plot(x.eval(), y.eval())
    axs[i].plot(x.eval(), z.eval())
    axs[i].plot(x.eval(), tf.nn.softmax(tf.concat((y, z), 1)).eval())
    axs[i].grid()
    axs[i].set_ylim(-1, 2)
```

```python
fig = plt.figure()
plt.plot(x.eval(), tf.nn.softmax(tf.ones((100))))
```
